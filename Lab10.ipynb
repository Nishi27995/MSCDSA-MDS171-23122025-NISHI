```python

class student:
    def __init__(self):
        self.details=[]
    def add_student(self,name,dep,regno,email,stats,maths,computer,research):
        det={
            'name':name,
            'dep':dep,
            'regno':regno,
            'email':email,
            'stats':stats,
            'maths':maths,
            'computer':computer,
            'research':research
        }
        self.details.append(det)
        print(f"{det}Added the Detais")
    def list_students(self):
        if self.details:
         print(self.details)
        else:
          print("Details not added")
    def search_student(self,regno):
        for regno in self.details:
            if regno == regno:
                print(regno)
                print("Student exist")
            else:
                print("Student dose not exist")
    def remove_student(self,regno):
        for regno in self.details:
            if regno==regno:
                self.details.remove(regno)
                print(self.details)
                print("Student details removed")
            else:
                print("Student dose not exist")
```
