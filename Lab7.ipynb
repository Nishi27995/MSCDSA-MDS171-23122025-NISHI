```python
import random

# Predefined menu and prices
menu = {
    "1": {"item": "Rice", "price": 250},
    "2": {"item": "Dal", "price": 200},
    "3": {"item": "Pizza", "price": 500},
    "4": {"item": "Palak Paneeer", "price": 250},
    "5": {"item": "Fried Rice", "price": 180},
    "6": {"item":"Masala Dosa","price": 70},
    "7": {"item": "Cake","price": 100},
    "8": {"item": "Ice-cream", "price": 60}
    
}

orders = {}

def generate_order_id():
    # Generate a unique orderID using 5 random numbers
    s=''
    for i in range(1,8):
        s=s+str(random.randint(0,9))
    return s

def create_new_order():
    order_id = generate_order_id()
    items = {}
    
    
    while True:
        item_no = input("Enter the item number to order (or 'done' to finish): ")
        
        if item_no == 'done':
            break
        
        if item_no in menu:
            quantity = int(input(f"Enter the quantity for {menu[item_no]['item']}: "))
            if item_no in items:
                items[item_no]['quantity'] += quantity
            else:
                items[item_no] = {'item': menu[item_no]['item'], 'quantity': quantity}
        else:
            print("Invalid item number. Please select a valid item.")
    
    orders[order_id] = items
    
    print(f"Order {order_id} created!")
def view_orders():
    for i, j in orders.items():
        print(f"Order {i}:")#we are extracting the order details under the respective order ids
        total_order_value = 0
        
        for k, l in j.items():#under item names and its respective quantity,we are multiplying the quantity with price
            item_name = l['item']
            quantity = l['quantity']
            price = menu[k]['price']
            item_total = price * quantity
            print(f"{item_name}: {quantity} x Rs.{price} = Rs.{item_total}")
            total_order_value += item_total
        
        print(f"Total Order Value: Rs.{total_order_value}")
        print("\n")
        print("_"*50)

# Main loop
print("Menu:")
print("_"*50)
for key, value in menu.items():#menu has keys as the item numbers and their respective values as item name and their prices
    print(f"{key}: {value['item']} - Rs.{value['price']}")
print("_"*50)
print("\n\n")

print("_"*50)
print("Options:")
print("1. Create new order")
print("2. View Orders")
print("3. Exit")
print("_"*50)
print("\n\n")
while True:
    
    choice = input("Enter your choice: ")
   
    print("choice entered : ", choice)
    print("\n")
    if choice == "1":
        create_new_order()
        print("_"*50)
    elif choice == "2":
        view_orders()
        print("_"*50)
    elif choice == "3":
        break
    else:
        print("Invalid choice. Please select a valid option.")
print("_"*50)
```
